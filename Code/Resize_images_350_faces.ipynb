The pictures of the 350 faces are of different sizes. Here we want to prepare them for annotation and resize them to the same size of 512x512 px.

```python
from PIL import Image
import os
from google.colab import drive
drive.mount('/content/drive')
```

We specify the paths to the pictures:

```python
input_directory = '/content/drive/MyDrive/ComputerVision/Data/Faces_350'
output_directory = '/content/drive/MyDrive/ComputerVision/Data/Faces_350_resized'
```

That are the files we want to resize:

```python
files = os.listdir(input_directory)
files
```

In the loop we resize every picture from the input directory and save it in the output directory:

```python
# Loop through each file in the input directory
for filename in files:
        # Open the image
        with Image.open(os.path.join(input_directory, filename)) as img:
            # Resize the image to 128 pixels
            img = img.resize((512,512), Image.ANTIALIAS)

            # Save the resized image to the output directory
            img.save(os.path.join(output_directory, filename))
```
